###  jpeg      /   ,    label  

```python
paths = []
dataset_gubuns = []
label_gubuns = []
# os.walk()          
# catanddog    jpg      
# catanddog  /train, /test/    (,     )

for dirname, _, filenames in os.walk('../../data/catanddog'):
    for filename in filenames:
        #       .
        if '.jpg' in filename:
            #    file_path  . 
            file_path = dirname+'/'+ filename
            paths.append(file_path)
            #    training_set, test_set      'train' 'test' . 
            if '\\training_set\\' in file_path:
                dataset_gubuns.append('train')  
            elif '\\test_set\\' in file_path:
                dataset_gubuns.append('test')
            else: dataset_gubuns.append('N/A')
            
            #    dogs     dog  , cats  cat  . 
            if 'dogs' in file_path:
                label_gubuns.append('DOG')
            elif 'cats' in file_path:
                label_gubuns.append('CAT')
            else: label_gubuns.append('N/A')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Preprocessing Data Loading  .   Preprocessing Data Loding Model fit_generator()     
# augmentation horizontal_flip( )  0~255 pixel 0~1 
train_gen = ImageDataGenerator(horizontal_flip=True, rescale=1/255.0)

# ImageDataGenerator  flow_from_diretory()  
# class_mode='categorical'  Label  - ,  array 224 x 224 ,  64 
train_flow_gen = train_gen.flow_from_directory(directory='../../data/catanddog/training_set/training_set', # image file  
                                               target_size=(224, 224), #    resize image size
                                               class_mode='categorical', #  label  one-hot encoding 
                                               batch_size=8) 
```

### Model 
* Input shape VGG, ResNet, Xception       224 x 224  .   
*   flow_from_directory()   224 x 224  .  

```python
IMAGE_SIZE = 224
BATCH_SIZE = 8
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense , Conv2D , Dropout , Flatten , Activation, MaxPooling2D , GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam , RMSprop 
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.callbacks import ReduceLROnPlateau , EarlyStopping , ModelCheckpoint , LearningRateScheduler
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications import Xception

def create_model(model_name='vgg16', verbose=False):
    
    input_tensor = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3))
    if model_name == 'vgg16':
        base_model = VGG16(input_tensor=input_tensor, include_top=False, weights='imagenet')
    elif model_name == 'resnet50':
        base_model = ResNet50V2(input_tensor=input_tensor, include_top=False, weights='imagenet')
    elif model_name == 'xception':
        base_model = Xception(input_tensor=input_tensor, include_top=False, weights='imagenet')
        
    bm_output = base_model.output
    
    x = GlobalAveragePooling2D()(bm_output)
    if model_name != 'vgg16':
        x = Dropout(rate=0.5)(x)
    x = Dense(50, activation='relu', name='fc1')(x)
    output = Dense(2, activation='softmax', name='output')(x)
    
    model = Model(inputs=input_tensor, outputs=output)
    
    if verbose:
        model.summary()
        
    return model
```

### optimizer, loss, metric 
*  flow_from_directory(class_mode='categorical')   label     
*  loss categorical_crossentropy   . 

```python
#  flow_from_directory(class_mode='categorical')   label     
# loss         loss  .  categorical_crossentropy 
#   class_mode='sparse'  loss sparse_categorical_crossentropy  .
model = create_model(model_name='xception', verbose=True)
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

###  fit_generator()   . 
* fit_generator()   iterator    image array, label array        .   
*   generator   image_data_generator 
* epochs       
* steps_per_epoch        . generator       batch               .   / ,    +1  .    1001 ,   10 steps_per_epochs 101 . 

```python
# train   generator samples     . 
train_image_cnt = train_flow_gen.samples
print(train_image_cnt)

# batch size 64,   image  8005
# model.fit_generator(train_flow_gen, epochs=15, 
#                     steps_per_epoch=int(np.ceil(train_image_cnt/BATCH_SIZE))
#                    )

#  fit() . (fit_generator()  )
model.fit(train_flow_gen, epochs=15, steps_per_epoch=int(np.ceil(train_image_cnt/BATCH_SIZE)))
```

### Test   
* Test   Generator . 

```python
# test data augmentation   . 
test_gen = ImageDataGenerator(rescale=1/255.)
test_flow_gen = test_gen.flow_from_directory(directory='../../data/catanddog/test_set/test_set' # image file   
                                                         ,target_size=(IMAGE_SIZE, IMAGE_SIZE) #    resize image size
                                                         ,class_mode='categorical' #  label  one-hot encoding . 
                                                         ,batch_size=BATCH_SIZE, shuffle=False
                                                         )
```

```python
#  evaluate_generator()  Test   .(evaluate() )
model.evaluate_generator(test_flow_gen)
```
