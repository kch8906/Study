### 중복순열 구하기

1부터 N까지 번호가 적힌 구슬이 있습니다. 이 중 중복을 허락하여 M번을 뽑아 일렬로 나열하는 방법을 모두 출력합니다.

- **출력설명**<br>
첫 번째 줄에 결과를 출력합니다. 맨 마지막 총 경우의 수를 출력합니다.<br>
출력순서는 사전순으로 오름차순으로 출력합니다.<br>
<br>
- **입력예제**<br>

```python
# 이렇게 하는거 아닌데 어떻게 풀어 나갈지 감도 안와서 구현만 해봄
def DFS(n1, n2):
    
    if n1 == n or n2 == n:       
        return
    else:     
        result.append((n1+1 , n2+1))
        DFS(n1 + 1, n2)
        DFS(n1, n2 + 1) 
        
n, m = map(int, input().split())
n_list = [0] * n
result = []

DFS(0, 0)

result = sorted(list(set(result)))
for i in result:
    for j in range(2):
        print(i[j], end=' ')
    print()
print(len(result))
```
