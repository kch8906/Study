## 03_Grouping

뉴욕 airBnB : https://www.kaggle.com/ptoscano230382/air-bnb-ny-2019<br> DataUrl = https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv

#### Question 44
- 데이터를 로드하고 상위 5개 컬럼을 출력하라

```python
Dataurl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df = pd.read_csv(Dataurl)
df.head(5)
```

#### Question 45
- 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
res = df.groupby('host_name')

# or

res = df['host_name'].value_counts().sort_index()
```

```python
res.head(5)
```

#### Question 46
- 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라

```python
res = df['counts'] = df.groupby('host_name')\
                        .size().to_frame().rename(columns={0:'counts'})\
                        .sort_values('counts', ascending=False)
res.head(5)
```

#### Question 47
- neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라

```python
res = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
res.head(5)
```

#### Question 48
- neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
res = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()\
        .groupby(['neighbourhood_group'], as_index=False).max()
res
```
